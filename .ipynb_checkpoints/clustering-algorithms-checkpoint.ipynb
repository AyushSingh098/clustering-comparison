```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn.metrics as sm
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
print(iris.target_names)
```

```python
x = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns=['Target'])
```

# Visualization

```python
plt.figure(figsize=(12,3))
colors = np.array(['red', 'green', 'blue'])
iris_targets_legend = np.array(iris.target_names)
red_patch = mpatches.Patch(color='red', label='Setosa')
green_patch = mpatches.Patch(color='green', label='Versicolor')
blue_patch = mpatches.Patch(color='blue', label='Virginica')


plt.subplot(1, 2, 1)
plt.scatter(x['Sepal Length'], x['Sepal Width'], c=colors[y['Target']])
plt.title('Sepal Length vs Sepal Width')
plt.legend(handles=[red_patch, green_patch, blue_patch])

plt.subplot(1,2,2)
plt.scatter(x['Petal Length'], x['Petal Width'], c= colors[y['Target']])
plt.title('Petal Length vs Petal Width')
plt.legend(handles=[red_patch, green_patch, blue_patch])
```

# Partitioning Approach

### 1. K-Means Clustering

```python
iris_k_mean_model = KMeans(n_clusters=3)
iris_k_mean_model.fit(x)
```

```python
print(iris_k_mean_model.cluster_centers_)
```

```python
plt.figure(figsize=(12,3))

colors = np.array(['red', 'green', 'blue'])

predictedY = np.choose(iris_k_mean_model.labels_, [1, 0, 2]).astype(np.int64)

plt.subplot(1, 2, 1)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[y['Target']])
plt.title('Before classification')
plt.legend(handles=[red_patch, green_patch, blue_patch])

plt.subplot(1, 2, 2)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[predictedY])
plt.title("Model's classification")
plt.legend(handles=[red_patch, green_patch, blue_patch])
```

```python
sm.accuracy_score(predictedY, y['Target'])
```

# Hierarchical approach

### 1. AGNES

```python
x = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns=['Target'])
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

# Hierarchical clustering
# Ward is the default linkage algorithm, so we'll start with that
ward = AgglomerativeClustering(n_clusters=3)
ward_pred = ward.fit_predict(x)
```

```python
# Hierarchical clustering using complete linkage
complete = AgglomerativeClustering(n_clusters=3,linkage='complete')
complete_pred = complete.fit_predict(iris.data)

# Hierarchical clustering using average linkage
avg = AgglomerativeClustering(n_clusters=3,linkage='average')
avg_pred = avg.fit_predict(iris.data)

# Hierarchical clustering using complete linkage
single = AgglomerativeClustering(n_clusters=3,linkage='single')
single_pred = single.fit_predict(iris.data)
```

```python
from sklearn.metrics import adjusted_rand_score

ward_ar_score = adjusted_rand_score(iris.target, ward_pred)
```

```python
# Calculated the adjusted Rand score for the complete linkage clustering labels
complete_ar_score = adjusted_rand_score(iris.target, complete_pred)

#Calculated the adjusted Rand score for the average linkage clustering labels
avg_ar_score = adjusted_rand_score(iris.target, avg_pred)

#Calculated the adjusted Rand score for the single linkage clustering labels
single_ar_score = adjusted_rand_score(iris.target, single_pred)
```

```python
print( "Scores: \nWard:", ward_ar_score,"\nComplete: ", complete_ar_score, "\nAverage: ", avg_ar_score, "\nSingle: ", single_ar_score)
```

```python
# Import scipy's linkage function to conduct the clustering
from scipy.cluster.hierarchy import linkage

# Specify the linkage type. Scipy accepts 'ward', 'complete', 'average', as well as other values

# Pick the one that resulted in the highest Adjusted Rand Score
linkage_type = 'single'

linkage_matrix = linkage(x, linkage_type)
```

#### (i) Single Linkage

```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
plt.figure(figsize=(11,9))

# plot using 'dendrogram()'
dendrogram(linkage_matrix)

plt.show()
```

#### (ii) Complete Linkage

```python
linkage_type = 'complete'

linkage_matrix = linkage(x, linkage_type)
```

```python
plt.figure(figsize=(11,9))

# plot using 'dendrogram()'
dendrogram(linkage_matrix)

plt.show()
```

# Density Based Clustering Method
### 1. DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
dbscan=DBSCAN()

#print(dbscan)

dbscan.fit(iris.data)

#labels
#print(dbscan.labels_)

pca=PCA(n_components=2).fit(iris.data)
pca_2d=pca.transform(iris.data)
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')
plt.legend([c1, c2, c3], ['Cluster 1', 'Cluster 2', 'Noise'])
plt.title('DBSCAN finds 2 clusters and nosie')
plt.show()

#two clusters have been identified, 0 and 1, while outlier observations are labeled -1.
```
